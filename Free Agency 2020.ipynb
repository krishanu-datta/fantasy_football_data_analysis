```python
## import relevant packages
import numpy as np
import pandas as pd
import urllib.request as urllib
import requests
from bs4 import BeautifulSoup
```

```python
## webscraping cell

## make dataframes to be filled

year = []
date = []
week= []
ttype = []
player = []
position = []
team = []
IR = []
start = []
finish = []
manager = []
device = []

## url setup
site1 = 'https://fantasy.nfl.com/league/2461806/history/'
site2= '/transactions?offset='

## loop from round 1 to round 16
for z in range(2020,2021):
    for y in range(1,2581,20):
        years = str(z)
        
        ## create soup object for every page of transactions
        site = site1 + years + site2 + str(y) 
        html = urllib.urlopen(site)
        soup = BeautifulSoup(html.read())
        
        ## isolate where the data is on the page
        transaction_table = soup.find("tbody")
        tr_tags = transaction_table.find_all("tr")
        
        ## go through the table of transactions
        for i in range(0,len(tr_tags)):
            ## add the year to the df
            year.append(int(z))
            
            ## find and add the date
            if tr_tags[i].find("td",{"class": "transactionDate"}) is not None:
                action_date = tr_tags[i].find("td",{"class": "transactionDate"}).get_text()
                date.append(action_date)
            else:
                date.append("N/A")
                
            ## find and add the week of the season
            if tr_tags[i].find("td",{"class": "transactionWeek"}) is not None:
                weekof = tr_tags[i].find("td",{"class": "transactionWeek"}).get_text()
                if (weekof != ""):
                    week.append(weekof)
                else:
                    week.append("0")
            else:
                week.append("0")
            
            ## find and add the transaction type
            if tr_tags[i].find("td",{"class": "transactionType"}) is not None:
                trtype = tr_tags[i].find("td",{"class": "transactionType"}).get_text()
                ttype.append(trtype)
            else:
                ttype.append("N/A")
            
            ## find and add player name
            if tr_tags[i].find("td",{"class": "playerNameAndInfo"}) is not None:
                player_card = tr_tags[i].find("td",{"class": "playerNameAndInfo"})
                if player_card.find("a",{"class": "playerName"}) is not None:
                    player_name = player_card.find("a",{"class": "playerName"}).get_text()
                    player.append(player_name)
                else:
                    player.append("N/A")
            else:
                player.append("N/A")
            
            
            ## find and add player position and team
            if tr_tags[i].find("em") is not None:
                player_position = tr_tags[i].find("em").get_text().split()[0]
                position.append(player_position)

                if len(tr_tags[i].find("em").get_text().split()) > 1:
                    real_life_team = tr_tags[i].find("em").get_text().split()[2]
                    team.append(real_life_team)

                else:
                    team.append("N/A")
            else:
                position.append("N/A")
                team.append("N/A")      
            
            ## find and add player IR status
            if tr_tags[i].find("strong") is not None:
                IR.append(tr_tags[i].find("strong").get_text())
            else:
                IR.append("N/A")
                
            ## find and add transaction origin
            if tr_tags[i].find("td",{"class": "transactionFrom"}) is not None:
                origin = tr_tags[i].find("td",{"class": "transactionFrom"}).get_text()
                start.append(origin)
            else:
                start.append("N/A")
                
             ## find and add transaction finish
            if tr_tags[i].find("td",{"class": "transactionTo"}) is not None:
                end = tr_tags[i].find("td",{"class": "transactionTo"}).get_text()
                finish.append(end)
            else:
                finish.append("N/A")
                
            ## find and add manager
            if tr_tags[i].find("td",{"class": "transactionOwner"}) is not None:
                ownercard = tr_tags[i].find("td",{"class": "transactionOwner"})
                owner = ownercard.find("span").get_text()
                manager.append(owner)
                if ownercard.find("a") is not None:
                    device.append("Phone")
                else:
                    device.append("Computer")
            else:
                manager.append("N/A")
                device.append("Computer")
            
```

```python
df = pd.DataFrame()
df["Year"] = year
df["Date"] = date
df["Week"] = week
df["Week"] = df["Week"].astype(float)
df["Type"] = ttype
df["Name"] = player
df["Position"] = position
df["Team"] = team
df["IR"] = IR
df["From"] = start
df["To"] = finish
df["Manager"] = manager
df["Device"] = device
```
